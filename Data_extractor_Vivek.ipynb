```python
import requests
from googlesearch import search
from urllib.parse import urlparse, urlunparse
import csv
from google.colab import files

def clean_url(url):
    parsed = urlparse(url)
    return urlunparse(parsed._replace(query="", fragment=""))

def is_shopify_site(url):
    try:
        headers = {"User-Agent": "Mozilla/5.0"}
        response = requests.get(url, headers=headers, timeout=10)
        content = response.text.lower()
        if "cdn.shopify.com" in content or "shopify.theme" in content or ".myshopify.com" in content:
            return True
        if any("shopify" in k.lower() or "shopify" in v.lower() for k, v in response.headers.items()):
            return True
        if "myshopify.com" in url.lower():
            return True
    except:
        pass
    return False

def is_wordpress_site(url):
    try:
        headers = {"User-Agent": "Mozilla/5.0"}
        response = requests.get(url, headers=headers, timeout=10)
        content = response.text.lower()
        if "/wp-content/" in content or "/wp-includes/" in content:
            return True
        if 'name="generator"' in content and "wordpress" in content:
            return True

        wp_json_url = url.rstrip('/') + "/wp-json/"
        wp_response = requests.get(wp_json_url, headers=headers, timeout=5)
        if wp_response.status_code == 200 and "wordpress" in wp_response.text.lower():
            return True

        if any("wordpress" in c.name.lower() for c in response.cookies):
            return True
    except:
        pass
    return False

output_file = "websites.csv"

def main():
    query = "jewelry stores in spain"
    print(f"Searching Google for: {query}")

    results = []
    for url in search(query, stop=300):
        results.append(clean_url(url))

    results = list(set(results))
    print(f"\nFound {len(results)} unique URLs\n")

    classified = []

    for site in results:
        try:
            if is_shopify_site(site):
                print(f"Shopify: {site}")
                classified.append([site, "Shopify"])
            elif is_wordpress_site(site):
                print(f"WordPress: {site}")
                classified.append([site, "WordPress"])
            #else:
                #print(f"Other: {site}")
                #classified.append([site, "Other"])
        except Exception as e:
            print(f"Error with {site}: {e}")
            classified.append([site, "Error"])


    with open(output_file, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["Website", "Platform"])
        writer.writerows(classified)

    print("\nCSV file saved")
    files.download(output_file)


main()
```
